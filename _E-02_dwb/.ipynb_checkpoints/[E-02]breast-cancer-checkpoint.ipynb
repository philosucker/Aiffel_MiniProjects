```python
# (1) 필요한 모듈 import 
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd

# (2) 데이터 준비
breast_cancer = load_breast_cancer()

# (3) 데이터 이해하기

# Feature Data 지정하기  (머신러닝 모델에게 입력되는 데이터를 feature 라고 부르기도 한다.)
breast_cancer_data = breast_cancer.data

# Label Data 지정하기
breast_cancer_label = breast_cancer.target

# Target Names 출력해 보기
print("Target_names :", breast_cancer.target_names, "\n")

# 데이터 Describe 해 보기
print("breast_cancer 데이터의 자료형 :", type(dir(breast_cancer)),"\n")
print("breast_cancer 데이터에 담긴 정보들 :", breast_cancer.keys(),"\n")
print("breast_cancer 데이터의 feature의 shape(총 데이터수, 각 데이터가 갖고 있는 정보의 수) :", breast_cancer_data.shape,"\n")
# print("breast_cancer 데이터의 feature의 맨 첫번째 데이터에 입력된 정보 :", breast_cancer_data[0]) 
print("breast_cancer 데이터의 총 label 개수 :", breast_cancer_label.shape,"\n")

# print(breast_cancer.DESCR)

breast_cancer_df = pd.DataFrame(data=breast_cancer_data, columns=breast_cancer.feature_names) 
breast_cancer_df["label"] = breast_cancer.target
print("breast_cancer 데이터의 각 feature와 target의 값 :")
breast_cancer_df
```

```python
# (4) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size=0.2, 
                                                    random_state= 25)
```

```python
# (4) 모델 학습 및 예측, 

# 첫번째 알고리즘 : Decision Tree 92%
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
# 두번째 알고리즘 : Random Forest 94%
from sklearn.ensemble import RandomForestClassifier


random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
# 세번째 알고리즘 : Support Vector Machine 91%
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
# 네번째 알고리즘 : Stochastic Gradient Descent Classifier (SGDClassifier)  92%

from sklearn.linear_model import SGDClassifier


sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
# 다섯번째 알고리즘 : Logistic Regression  90%

from sklearn.linear_model import LogisticRegression
import numpy as np


print('최소값:',np.min(X_train), ' 최대값:',np.max(X_train))
print('최소값:',np.min(X_test), ' 최대값:',np.max(X_test))
```

```python
# X_train 과 X_test Normalization
X_train_norm = X_train/255.0 
X_test_norm = X_test/255.0 

print('최소값:',np.min(X_train_norm), ' 최대값:',np.max(X_train_norm))
print('최소값:',np.min(X_test_norm), ' 최대값:',np.max(X_test_norm))

logistic_model = LogisticRegression()

logistic_model.fit(X_train_norm, y_train)
y_pred = logistic_model.predict(X_test_norm)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
유방암 분류 문제에서 가장 좋은 성능을 낸 모델 : Random forest 
```

```python
우선 성능지표 : 레이블 0에서의 Recall점수

근거
각 모델의 classification report를 보면
악성(malignant) 레이블이 0, 양성(benign)레이블이 1이다. 
실제로 악성인 사람 중에 모델이 악성이라고 분류한 비율인 Recall이
모델이 악성이라고 분류한 사람중에 실제로 악성인 사람의 비율인 Precision보다 높아야 한다.
모델이 양성을 악성으로 잘못 분류(FP)한 것이 악성을 양성으로 잘못 분류(FN)한 것보다 낫기 때문이다.

따라서 레이블 0에서 recall이 가장 높게 나온 모델인 Random forest가 최선의 모델이라 할 수 있다.
```
