```python
# (1) 필요한 모듈 import 
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd

# (2) 데이터 준비
digits = load_digits()

# (3) 데이터 이해하기

# Feature Data 지정하기  (머신러닝 모델에게 입력되는 데이터를 feature 라고 부르기도 한다.)
digits_data = digits.data

# Label Data 지정하기
digits_label = digits.target

# Target Names 출력해 보기
print("Target_names :", digits.target_names, "\n")

# 데이터 Describe 해 보기
print("digits 데이터의 자료형 :", type(dir(digits)),"\n")
print("digits 데이터에 담긴 정보들 :", digits.keys(),"\n")
print("digits 데이터의 feature의 shape(총 데이터수, 각 데이터가 갖고 있는 정보의 수) :", digits_data.shape,"\n")
# print("digits 데이터의 feature의 맨 첫번째 데이터에 입력된 정보 :", digits_data[0]) # 하나의 픽셀당 0 ~ 15까지의 숫자 정보
print("digits 데이터의 총 label 개수 :", digits_label.shape,"\n")

# print(digits.DESCR)

digits_df = pd.DataFrame(data=digits_data, columns=digits.feature_names) 
digits_df["label"] = digits.target
print("digits 데이터의 각 feature와 target의 값 :")
digits_df
```

```python
# (4) train, test 데이터 분리
X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label, 
                                                    test_size=0.2, 
                                                    random_state= 25)
```

```python
# (4) 모델 학습 및 예측, 

# 첫번째 알고리즘 : Decision Tree 84%
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
# 두번째 알고리즘 : Random Forest 97%
from sklearn.ensemble import RandomForestClassifier


random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
# 세번째 알고리즘 : Support Vector Machine 99%
from sklearn import svm

svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
# 네번째 알고리즘 : Stochastic Gradient Descent Classifier (SGDClassifier)  93%

from sklearn.linear_model import SGDClassifier


sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
# 다섯번째 알고리즘 : Logistic Regression  91%

from sklearn.linear_model import LogisticRegression
import numpy as np


print('최소값:',np.min(X_train), ' 최대값:',np.max(X_train))
print('최소값:',np.min(X_test), ' 최대값:',np.max(X_test))
```

```python
# X_train 과 X_test Normalization
X_train_norm = X_train/255.0 
X_test_norm = X_test/255.0 

print('최소값:',np.min(X_train_norm), ' 최대값:',np.max(X_train_norm))
print('최소값:',np.min(X_test_norm), ' 최대값:',np.max(X_test_norm))

logistic_model = LogisticRegression()
#logistic_model = LogisticRegression(solver='lbfgs', max_iter=10000).fit(X_train, y_train)

logistic_model.fit(X_train_norm, y_train)
y_pred = logistic_model.predict(X_test_norm)

print("confusion_matrix : ","\n", confusion_matrix(y_test, y_pred), "\n")
print("classification_report : ","\n", classification_report(y_test, y_pred), "\n")
```

```python
손글씨 이미지 분류 문제에서 가장 좋은 성능을 낸 모델 : SVM
```

```python


이 문제는 총 1797개의 숫자 이미지를 10개의 카테고리로 분류하는 최고의 알고리즘을 찾는 문제다.
이를 위해 각 이미지 당 64개의 feature가 입력돼있다.
서포트 벡터 머신은 99%의 성적을 냈다.
10개의 카테고리에서 각각 precision, recall, f1-score를 비교하는 게 무의미할 정도로 좋은 성적을 냈다.
따라서 이 문제에 있어서는 서포트 벡터 머신이 최선의 모델이라고 할 수 있다

그렇다면 왜 SVM이 최고의 결과를 낼 수 있었던 걸까?
레이블의 종류가 많고, feature의 개수도 많을 때는 SVM이 가장 좋은 모델이라고 추론해볼 수 있다.
```
