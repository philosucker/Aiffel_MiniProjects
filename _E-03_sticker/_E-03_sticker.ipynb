# 이미지 준비

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
my_image_path = os.getenv('HOME') + '/mini_projects/_E-03_sticker/images/image.jpg'
img_bgr = cv2.imread(my_image_path) #open CV는 BGR순으로 이미지 채널을 읽음. 그래서 파일이름을 bgr
img_bgr = cv2.resize(img_bgr, (960, 1280))
img_show = img_bgr.copy()  # 출력용 이미지 img_show
plt.imshow(img_bgr)
plt.show()
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # open CV는 BGR순으로 이미지 채널을 읽음. plt, dlib는 RGB순.
plt.imshow(img_rgb)
plt.show()   # <<<< 이 코드가 없어도 사진은 나오는데 뭔 차이지?
```

# Face Detection

```python
import dlib
detector_hog = dlib.get_frontal_face_detector()   # hog detector 선언
```

```python
dlib_rects = detector_hog(img_rgb, 1) # 두번째 파라미터는 이미지 피라미드의 숫자. 
print(dlib_rects)  # 찾은 얼굴영역 좌표
```

```python
for dlib_rect in dlib_rects:
    l = dlib_rect.left()
    t = dlib_rect.top()
    r = dlib_rect.right()
    b = dlib_rect.bottom()
    
    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # 출력용 이미지 img_show 에 boundingbox 작업
```

```python
img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)  # 출력용 이미지 img_show  채널 변경
plt.imshow(img_show_rgb)
plt.show()
```

# Face Landmark

```python
model_path = os.getenv('HOME') + '/mini_projects/_E-03_sticker/models/shape_predictor_68_face_landmarks.dat'
landmark_predictor = dlib.shape_predictor(model_path)
```

```python
list_landmarks = []
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_rgb, dlib_rect)
    list_points = list(map(lambda p:(p.x, p.y), points.parts()))
    list_landmarks.append(list_points)
print(len(list_landmarks[0]))
```

```python
for landmark in list_landmarks:
    for idx, point in enumerate(list_points):
        cv2.circle(img_show, point, 2, (0, 255, 255), -1) #yellow

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show
```

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks):
    print (landmark[33])
    x = landmark[33][0] - dlib_rect.height()//2 # 여기가 핵심
    y = landmark[33][1] - dlib_rect.width()//2
    w = dlib_rect.width()
    h = dlib_rect.width()
    print('(x,y) : (%d,%d)'%(x,y))
    print('(w,h) : (%d,%d)'%(w,h))
```

# 고양이 이미지 준비


```python
sticker_path = os.getenv('HOME') + '/mini_projects/_E-03_sticker/images/cat-whiskers.png'
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print(img_sticker.shape)
plt.imshow(img_sticker)
plt.show()
```

# 얼굴 사진에 고양이 이미지 적용

```python
sticker_area = img_show[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]]
img_show[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]] = \
np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8) #왜 0에서 255로 바꿔야 되는 건가?
print(img_sticker.shape[0])
```

```python
plt.imshow(cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB))
plt.show
```

# 바운딩 박스, 랜드마크 제거

```python
sticker_area = img_bgr[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]]
img_bgr[y:y+img_sticker.shape[0], x:x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8) #왜 0에서 255로 바꿔야 되는 건가?
plt.imshow(cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB))
plt.show()
```

## 프로젝트 진행 중 겪었던 문제점과 의문

**1. face detection 단계에서**   
   이미지의 height가 width 보다 작은 경우, 그러니까 사진이 위아래로 눌려서 얼굴이 찐빵 같이 로드됐을 때   
   bounding box를 잡지 못하는 문제가 발견되었다.   
      >> height를 width보다 늘려서 비율을 맞춰주니 해결되었다.   
      그런데 왜 가로 세로 비율이 맞지 않는 사진의 경우엔 bounding box를 잡지 못했던 것일까?   
      >>>dlib를 사용해서 얼굴좌표를 반환하는 객체의 함수에 비밀이 숨어 있는 것 같다   
         
   그렇다면 짜부러지거나 늘어난 사진에도 bounding box를 잡을 수 있게 하려면 어떻게 해야 할까?   
   
**2. face landmark 단계에서 고양이 수염을 얼굴에 붙일 때**   
    x와 y의 랜드마크 값을, 노드의 가이드에선 수염을 볼에 붙이기 위해 다양한 랜드마크값을 참조하라고 되어 있었는데   
    고양이 수염의 경우 고양이 코를 얼굴사진의 코에 붙이는 게 가장 효율적이라고 생각해 별도의 계산과정 없이   
    코 랜드마크[30]에 바로 적용했다.   
    그런데   
    아래와 같이 할 때와   
    x = landmark[30][0] - dlib_rect.height()//2     
    y = landmark[30][1] - dlib_rect.width()//2   
    아래와 같이 할 때   
    x = landmark[30][0]    
    y = landmark[30][1]    
    동일한 결과를 얻을 수 있었다   
    왜?   
    
**3. 얼굴 사진에 고양이 이미지 적용 단계에서**   
   처음에 고양이 수염을 붙이면 고양이 이미지에서 수염을 제외한 배경 부분이 하얗게 나와 얼굴을 가렸다.
   np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)   
      >> 위 코드에서 원래 img_sticker==0 이었던 것을 위와 같이 바꾸었더니 문제가 해결되었다.   
      >> 0과 255인걸 보면 색상정보와 관련이 있는 것 같은데, 0과 255가 아닌 다른 숫자를 넣었을 땐 문제가 해결되지 않았다.   
      >> 왜 그런거지?
